```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# read in results from experiment 1
exp1_results = pd.read_csv('C:/6120-NLP/project/minitagger-output/exp1-results.csv')
exp1_results.head()
```

```python
# create plot
plt.plot( 'Number of Labels', 'Baseline (Platt)', data=exp1_results, marker='o', markerfacecolor='brown', 
         color='brown', linewidth=2)
plt.plot( 'Number of Labels', 'Baseline + Bitstrings (Platt)', data=exp1_results, marker='o', 
         markerfacecolor='blue', color='blue', linewidth=2)
plt.plot( 'Number of Labels', 'Baseline (S&C)', data=exp1_results, marker='o', 
         markerfacecolor='brown', color='brown', linewidth=2, linestyle='dashed')
plt.plot( 'Number of Labels', 'Baseline + Bitstrings (S&C)', data=exp1_results, marker='o', 
         markerfacecolor='blue', color='blue', linewidth=2, linestyle='dashed')
# show legend
plt.rcParams["figure.figsize"] = (7,5)
plt.rcParams.update({'font.size': 12})
plt.xlabel('NUMBER OF LABELED WORDS') 
plt.ylabel('DEV ACCURACY') 
plt.legend()
plt.savefig('exp1.png', facecolor='w')
```

```python
# read in results from experiment 2
exp2_results = pd.read_csv('C:/6120-NLP/project/minitagger-output/exp2-results.csv')
exp2_results.head()
```
